# Python Data Cleaning Practice
Working with transactions dataset

```python
transactions = [
    100, -50, None, 200, "300", 100, -10, "invalid", 400,
    "500.5", 0, "", "700", -300, 200, "NULL"
]
```

```python
"""
 Q1. Remove Missing / Null-like Values
👉 Remove:
None
""
"NULL"

Pattern: Data cleaning (standardization)
"""

cleaned = [x for x in transactions if x not in (None, "", "NULL")]
cleaned
```

```python
"""
Q2. Standardize Data Types
👉 Convert:
"300" → 300
"500.5" → 500.5
👉 Ignore:
"invalid"

Pattern: Transformation + validation
"""

standardized = []

for x in transactions:
    if x in (None, "", "NULL"):
        continue

    if isinstance(x, str):
        try:
            if "." in x:
                x = float(x)
            else:
                x = int(x)
        except:
            continue

    standardized.append(x)

standardized
```

```python
"""
Q3. Keep Only Valid Numeric Records
👉 Final list should contain only:
int and float

Pattern: Data validation
"""

valid_numbers = []

for x in transactions:
    if isinstance(x, (int, float)):
        valid_numbers.append(x)
    elif isinstance(x, str):
        try:
            if "." in x:
                valid_numbers.append(float(x))
            else:
                valid_numbers.append(int(x))
        except:
            pass

valid_numbers
```

```python
"""
Q4. Apply Business Rule Filtering
👉 Keep only values: Greater than 0

Pattern: Business filtering logic
"""
filtered = []

for x in transactions:
    try:
        if isinstance(x, str):
            x = float(x) if "." in x else int(x)

        if isinstance(x, (int, float)) and x > 0:
            filtered.append(x)
    except:
        pass

filtered
```

```python
"""
Q5. Deduplicate Records
👉 Remove duplicates after cleaning
Pattern: Data deduplication
"""
unique = []

for x in transactions:
    try:
        if isinstance(x, str):
            x = float(x) if "." in x else int(x)

        if isinstance(x, (int, float)) and x > 0:
            if x not in unique:
                unique.append(x)
    except:
        pass

unique
```

```python
"""
Q6. Compute Total Transaction Value
👉 Sum all cleaned values
Pattern: Aggregation
"""
total = 0

for x in transactions:
    try:
        if isinstance(x, str):
            x = float(x) if "." in x else int(x)

        if isinstance(x, (int, float)) and x > 0:
            total += x
    except:
        pass

total
```

```python
"""
Q7. Count Records by Type
👉 Count:
how many integers
how many floats

Pattern: Basic data profiling
"""
int_count = 0
float_count = 0

for x in transactions:
    try:
        if isinstance(x, str):
            x = float(x) if "." in x else int(x)

        if isinstance(x, int):
            int_count += 1
        elif isinstance(x, float):
            float_count += 1
    except:
        pass

int_count, float_count
```

```python
"""
Q8. Identify Invalid Records (Important ⭐)
👉 Create a separate list of: invalid values ("invalid", "NULL", etc.)
Pattern: Data quality tracking
"""
invalid = []

for x in transactions:
    if x in (None, "", "NULL"):
        invalid.append(x)
    elif isinstance(x, str):
        try:
            float(x)
        except:
            invalid.append(x)

invalid
```

```python
"""
Q9. Sort Clean Data
👉 Sort final cleaned data in ascending order
Pattern: Data preparation for reporting
"""
clean = []

for x in transactions:
    try:
        if isinstance(x, str):
            x = float(x) if "." in x else int(x)

        if isinstance(x, (int, float)) and x > 0:
            if x not in clean:
                clean.append(x)
    except:
        pass

sorted(clean)
```

```python
"""
Q10. Build Mini Data Processing Function ⭐⭐
👉 Write a function:
def clean_transactions(data):
    return cleaned_data
👉 It should:
Handle nulls
Convert values
Remove invalid
Apply business rules
Deduplicate
"""
def clean_transactions(data):
    result = []

    for x in data:
        if x in (None, "", "NULL"):
            continue

        try:
            if isinstance(x, str):
                x = float(x) if "." in x else int(x)
        except:
            continue

        if isinstance(x, (int, float)) and x > 0:
            if x not in result:
                result.append(x)

    return result


transactions = [
    100, -50, None, 200, "300", 100, -10, "invalid", 400,
    "500.5", 0, "", "700", -300, 200, "NULL"
]

clean_transactions(transactions)
```
