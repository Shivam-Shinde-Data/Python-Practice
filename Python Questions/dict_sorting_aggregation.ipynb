# April 23/04 - Dict sorting and aggregation

```python
orders = [
    {"order_id": 1, "customer": "A", "amount": 100, "city": "Pune"},
    {"order_id": 2, "customer": "B", "amount": 200, "city": "Mumbai"},
    {"order_id": 3, "customer": "A", "amount": 150, "city": "Pune"},
    {"order_id": 4, "customer": "C", "amount": 300, "city": "Delhi"},
    {"order_id": 5, "customer": "B", "amount": 50, "city": "Mumbai"},
    {"order_id": 6, "customer": "A", "amount": 200, "city": "Pune"},
]
```

```python
"""
Q1. Sort Orders by Amount (Descending)
👉 Return full list sorted by amount (highest first)
Pattern: Sorting with key
"""
sorted_orders = sorted(orders, key=lambda x: x["amount"], reverse=True)
print(sorted_orders)
```

```python
"""
Q2. Top 2 Highest Orders 
👉 Return top 2 orders based on amount
Pattern: Sort + slice
"""
top_2 = sorted(orders, key=lambda x: x["amount"], reverse=True)[:2]
print(top_2)
```

```python
"""
Q3. Second Highest Order 
👉 Return only the second highest order
👉 Edge case: if same values exist?
Pattern: Sorting + indexing
"""
sorted_orders = sorted(orders, key=lambda x: x["amount"], reverse=True)
second_highest = sorted_orders[1]
print(second_highest)
```

```python
"""
Q4. Group Orders by Customer (Structured Output) 
👉 Output:
{
    "A": [100, 150, 200],
    "B": [200, 50],
    "C": [300]
}
👉 Only store amounts (not full dict)
Pattern: Grouping + transformation
"""
result = {}

for o in orders:
    if o["customer"] not in result:
        result[o["customer"]] = []
    
    result[o["customer"]].append(o["amount"])

print(result)
```

```python
"""
Q5. Customer-wise Sorted Amounts ⭐⭐
👉 Output:
{
    "A": [200, 150, 100],
    "B": [200, 50],
    "C": [300]
}
👉 Sort each customer’s data (descending)
Pattern: Group + sort
"""
result = {}

for o in orders:
    if o["customer"] not in result:
        result[o["customer"]] = []
    
    result[o["customer"]].append(o["amount"])

for key in result:
    result[key].sort(reverse=True)

print(result)
```

```python
"""
Q6. Filter High-Value Orders
👉 Return orders where: amount > 150
Pattern: Filtering with condition
"""
result = []

for o in orders:
    if o["amount"] > 150:
        result.append(o)

print(result)
```
